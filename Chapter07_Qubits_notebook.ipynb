# Chapter 7: Qubits

Corresponding to the classical bits 0 and 1, quantum gate computing works with the
computational basis qubits $|0\rangle$ and $|1\rangle$ (pronounced *ket 0* and *ket 1*), in
Dirac notation. A qubit is any $|q\rangle = x|0\rangle + y|1\rangle$ whose complex probability
amplitudes satisfy $|x|^2+|y|^2=1$; unlike a classical bit it may live in a *superposition* of
$|0\rangle$ and $|1\rangle$, and multiple qubits combine into larger systems such as
$|00\rangle,|01\rangle,|10\rangle,|11\rangle$. Because a measurement collapses a qubit to 0 or 1
with probabilities $|x|^2$ and $|y|^2$ (the Born rule), a state can only be estimated
statistically from many reads.

This notebook demonstrates the mechanics in Qiskit: constructing single- and multi-qubit states
with `Statevector`, checking normalization, displaying them in Dirac notation, forming inner
products, and building larger systems with the tensor product.

---


**Prerequisites:**
- See `Chapter02_QuantumSoftware.ipynb` for installation instructions


```python
# Setup and imports
import numpy as np
from IPython.display import display
from qiskit.quantum_info import Statevector
```

## Creating and displaying qubits  *(Book §7.1, §7.4, Listing 7.1)*

Using the vector representation $|0\rangle\equiv\begin{bmatrix}1\\0\end{bmatrix}$,
$|1\rangle\equiv\begin{bmatrix}0\\1\end{bmatrix}$, we build states directly as `Statevector`
objects: a single-qubit superposition $\tfrac{1}{\sqrt2}|0\rangle+\tfrac{i}{\sqrt2}|1\rangle$ and a
two-qubit state. `draw('latex')` renders them in Dirac notation, and `is_valid()` confirms the
amplitudes are normalized ($|x|^2+|y|^2=1$), i.e. that the object is a legal qubit system.


```python
ket0 = Statevector([1, 0]) # define ket 0
ket1 = Statevector([0, 1]) # define ket 1

phi = 1/np.sqrt(2)*ket0 +   1j/np.sqrt(2)*ket1
display(phi.draw('latex'))

psi =  Statevector([1/np.sqrt(2), 1j/np.sqrt(2),0,0]) 
display(psi.draw('latex'))

print(psi.is_valid())
```

## Creating kets using labels  *(Book §7.4, Listing 7.2)*

A convenient alternative is `Statevector.from_label`, which builds a basis state straight from a bit
string (`'0'`, `'1'`, `'00'`, `'11'`, ...); these combine naturally into superpositions such as
$|\phi\rangle=\tfrac12|00\rangle+\tfrac{\sqrt3}{2}|11\rangle$. When a multi-qubit state is stored as
a vector, the components correspond to basis states in binary counting order
($|00\rangle,|01\rangle,|10\rangle,|11\rangle$).


```python
ket0 = Statevector.from_label('0')# define ket 0
ket1 = Statevector.from_label('1') # define ket 1

phi = (1/2)*Statevector.from_label('00') + (np.sqrt(3)/2)*Statevector.from_label('11')
display(phi.draw('latex'))
```

## Inner products of qubits  *(Book §7.5, Examples 7.12–7.13)*

The *bra* $\langle\psi|$ is the conjugate transpose (adjoint) of the *ket* $|\psi\rangle$, and
together they form Dirac's bra-ket notation. The inner product is
$\langle\phi|\psi\rangle=\sum_i y_i^{*}x_i$, so the basis kets are orthonormal
($\langle 0|1\rangle=0$) and, because qubit states are normalized, $\langle\psi|\psi\rangle=1$.
Here `psi.inner(phi)` evaluates this overlap for two two-qubit states; note
$\langle\phi|\psi\rangle=\langle\psi|\phi\rangle^{*}$.


```python
psi =  Statevector([-1j/np.sqrt(2), 0,0, 1/np.sqrt(2)]) 
phi =  Statevector([-1j/np.sqrt(2), (np.sqrt(3)-1j)/4,0, 1/2]) 
display(psi.draw('latex'))
display(phi.draw('latex'))
print(psi.inner(phi))
```

## Tensor products of qubits  *(Book §7.6, Listing 7.3, Example 7.14)*

The tensor product $\otimes$ joins separate qubits into one system:
$|q_1\rangle\otimes|q_0\rangle\equiv|q_1 q_0\rangle$, and in general
$|q_{n-1}\rangle\otimes\cdots\otimes|q_0\rangle\equiv|q_{n-1}\cdots q_0\rangle$. In Qiskit,
`ket0.tensor(ket1)` builds $|01\rangle$ from the single-qubit kets; we assemble a two-qubit
superposition from these building blocks and then tensor it with itself to obtain a four-qubit
state.


```python
ket0 = Statevector([1, 0]) # define ket 0
ket1 = Statevector([0, 1]) # define ket 1

ket00 = ket0.tensor(ket0)
ket01 = ket0.tensor(ket1)
ket10 = ket1.tensor(ket0)
ket11 = ket1.tensor(ket1)

phi = 1/2*ket00 +   1j/np.sqrt(2)*ket10 +  (np.sqrt(3)+1j)/4*ket11

display(phi.draw('latex'))
psi = phi.tensor(phi)
display(psi.draw('latex'))
```
