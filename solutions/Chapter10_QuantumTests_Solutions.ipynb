## Truss displacement inner product

```python
def quantumPadding(v):
    # Pad the vector to length 2^m
    v_flat = v.flatten()
    m = int(np.ceil(np.log2(len(v_flat))))
    padded = np.pad(v_flat, (0, 2**m - len(v_flat)))
    nrm = np.linalg.norm(padded)
    return padded / nrm, nrm
    
fem_model= truss3x3(A=0.005)
d_A, valid = fem_model.solve()
fem_model.A[0] = 0.0025
fem_model.A[1] = 0.0025
d_B, valid = fem_model.solve()
# Classical inner product
classical_ip = np.dot(d_A, d_B)
print(f"Classical inner product: {classical_ip:.6e}")

# Quantum inner product using SWAP test
psi_A, nrm_A = quantumPadding(d_A)
psi_B, nrm_B = quantumPadding(d_B)

quantum_ip, qc = innerProductEstimation(psi_A, psi_B, shots=1000)
quantum_ip *= nrm_A * nrm_B  # Scale back by the norms
print(f"Quantum inner product estimate: {quantum_ip:.6e}")

```
