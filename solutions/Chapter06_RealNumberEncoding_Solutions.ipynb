## Using 3 qubits per variable

```python
# Problem data
A = np.array([[4, -1], [-1, 2]])
b = np.array([5.5, 0.5])
x_exact = np.linalg.solve(A, b)
print("Exact solution:", x_exact)
N = 2

# Box parameters
c = np.array([0, 0])
L = 0.5

# Create binary variables
q1 = Array.create('q1', shape=N, vartype='BINARY')
q2 = Array.create('q2', shape=N, vartype='BINARY')
q3 = Array.create('q3', shape=N, vartype='BINARY')

# Box representation: x = c + L*(-4*q1 + 2*q2 + q3)
x = [c[i] + L*(-4*q1[i] + 2*q2[i] + q3[i]) for i in range(N)]
print(x)
# Potential energy: Pi = 0.5*x^T*A*x - x^T*b
Pi = 0
for i in range(N):
    for j in range(N):
        Pi += 0.5 * A[i,j] * x[i] * x[j]
    Pi -= b[i] * x[i]

# Compile and solve
model = Pi.compile()
bqm = model.to_bqm()


sampler = ExactSolver()
results = sampler.sample(bqm)
q, HMin = print_QUBO_best_result(results,"Exact: ")
# Extract best solution
# Map from q to x using the best solution q

x_best = np.array([c[i] + L*(-4*q[i] + 2*q[i+N] + q[i+2*N]) for i in range(N)])
print("Best x:", x_best)
```
